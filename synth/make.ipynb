```python
import matplotlib.animation as anime
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import random

def func(x,var):
	A=var*random.random()
	B=4-8*random.random()
	C=5-8*random.random()
	return A*np.sin(B*x+C)

def rand_func(var,bias):
	xlist=[]
	ylist=[]
	for xval in np.linspace(-10, 10, 100):
		xlist.append(xval)
		temp_y=bias
		for _ in range(0,8):
			temp_y+=func(xval,var)
		ylist.append(temp_y)
	return xlist, ylist
```

```python
fig,ax = plt.subplots(dpi=300)


# p1, = plt.plot([], [], 'ko')
# p2, = plt.plot([], [], 'mo')

plt.xlim(-10, 10)
plt.ylim(-20, 20)

xlist, ylist=rand_func(2,-8)
xlist2,ylist2=rand_func(4,4)
max1=np.full_like(xlist,max(ylist))
min1=np.full_like(xlist,min(ylist))
max2=np.full_like(xlist,max(ylist2))
min2=np.full_like(xlist,min(ylist2))
l, = plt.plot(xlist, ylist, 'k-')
l2, = plt.plot(xlist2, ylist2, 'r-')
m1, = plt.plot(xlist,max1, 'k--',alpha=0.5)
m2, = plt.plot(xlist,min1, 'k--',alpha=0.5)
m3, = plt.plot(xlist,max2, 'r--',alpha=0.5)
m4, = plt.plot(xlist,min2, 'r--',alpha=0.5)
ax.fill_between(xlist, max2, min2, interpolate=True, color='red', alpha=0.2)
ax.fill_between(xlist, max1, min1, interpolate=True, color='grey', alpha=0.5)

```
